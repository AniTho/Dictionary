```python
import json
from difflib import SequenceMatcher
```

```python
file=json.load(open('data.json.json'))
```

```python
word=input("Enter word you want to search: ").lower()
```

```python
def Output(word,NewDict):
    listOfMeaning=NewDict[word]
    j=1
    for i in listOfMeaning:
        print(str(j)+".) "+i)
        j+=1
```

```python
flag=0
NewDict={k.lower(): v for k, v in file.items()}
if(word in NewDict):
    Output(word,NewDict)
    flag=1

else: 
    for i in NewDict:    
        if(SequenceMatcher(None,word,i).ratio()>0.75):
            check=input("Do you mean: "+i+"\nEnter Y for yes and N for no ")
            
            if(check=='Y' or check=='y'):
                Output(word,NewDict)
                flag=1
                break
            
            else:
                continue
        
         
        
if(flag==0):
    print("Sorry No word found")
```
